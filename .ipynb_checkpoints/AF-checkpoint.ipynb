```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
def deriv_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
x=np.linspace(-10,10,100)
y=sigmoid(x)
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
```

```python
z=deriv_sigmoid(x)
plt.plot(x,z)
```

```python
def tanh(x):
    return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
    
```

```python
def derive_tanh(x):
    return 1-tanh(x)**2
```

```python
x=np.linspace(-10,10,100)
y=tanh(x)
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=derive_tanh(x)
plt.plot(x,y)
```

```python
def relu(x):
    return np.maximum(0,x)
def derive_relu(x):
    return np.where(x>0,1,0)
```

```python
x=np.linspace(-10,10,100)
y=relu(x)
plt.plot(x,y)
```

```python
z=derive_relu(x)
plt.plot(x,z)
```

```python
def lrelu(x,alp=0.1):
    return np.maximum(x,alp*x)
def derive_lrelu(x,alp=0.1):
    return np.where(x>0,1,alp)
```

```python
y=lrelu(x)
plt.plot(x,y)
```

```python
z=derive_lrelu(x)
plt.plot(x,z)
```

```python
def elu(x,alp=1):
    return np.where(x>0,x,alp*(np.exp(x)-1))
y=elu(x)
plt.plot(x,y)
```

```python
def derive_elu(x,alp=1):
    return np.where(x>0,1,alp*np.exp(x))
z=derive_elu(x)
plt.plot(x,z)
```

```python
def softplus(x):
    return np.log(1+np.exp(x))
x=np.linspace(-10,10,100)
y=softplus(x)
plt.plot(x,y)
```

```python
def derive_softplus(x):
    return 1/(1+np.exp(-x))
z=derive_softplus(x)
plt.plot(x,z)
```

```python
def softmax(x):
    x_shifted = x - np.max(x, axis=-1, keepdims=True)  # Stability trick
    exp_x = np.exp(x_shifted)
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
```

```python
y=softmax(x)
plt.plot(x,y)
```
