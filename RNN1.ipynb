```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,Dropout
from tensorflow.keras import Input,Model
```

```python
dataset_path = "SMSSpamCollection"
df = pd.read_csv(dataset_path, sep='\t', header=None, names=['label', 'message'], encoding='utf-8')
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
max_words=5000
tokenizer=Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df['message'])
sequences=tokenizer.texts_to_sequences(df['message'])
max_length=max([len(x) for x in sequences])
padded=pad_sequences(sequences,maxlen=max_length)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(padded,df['label'],test_size=0.2,random_state=42)
```

```python
vocal_length=min(len(tokenizer.word_index)+1,max_words)
```

```python
inputs=Input(shape=(max_length,))
x=Embedding(vocal_length,128)(inputs)
x=SimpleRNN(64,return_sequences=True)(x)
x=SimpleRNN(32)(x)
x=Dense(16,activation='relu')(x)
x=Dropout(0.5)(x)
outputs=Dense(1,activation='sigmoid')(x)
model=Model(inputs=inputs,outputs=outputs)
```

```python
model.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
history=model.fit(xtrain,ytrain,epochs=10,batch_size=32,validation_data=(xtest,ytest),validation_split=0.2)
```

```python
loss,acc=model.evaluate(xtest,ytest)
```

```python
ypred=(model.predict(xtest)>0.5).astype("int32")
print(accuracy_score(ytest,ypred))
```

```python
def predict_rnn(message):
    seq=tokenizer.texts_to_sequences(message)
    pad=pad_sequences(seq,maxlen=max_length,padding='post')
    ypred=model.predict(pad)[0][0]
    return "spam" if ypred>0.5 else "ham"
```

```python
print(predict_rnn("Congratulations! You've won a free gift."))
```
