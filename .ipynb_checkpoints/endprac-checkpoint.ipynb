CNN mnist

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
(xtrain,ytrain),(xtest,ytest)=datasets.mnist.load_data()
```

```python
xtrain,xtest=xtrain/255.0,xtest/255.0
```

```python
model=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128,activation='relu'),
    Dense(10,activation='softmax')
    
])
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
earlystopping=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
history=model.fit(xtrain,ytrain,epochs=10,validation_data=(xtest,ytest),callbacks=[earlystopping])
```

```python
loss,acc=model.evaluate(xtest,ytest)
```

```python
acc
```

```python
ypred=model.predict(xtest)
print(classification_report(ytest,np.argmax(ypred,axis=1)))
```

```python
cm=confusion_matrix(ytest,np.argmax(ypred,axis=1))
p=ConfusionMatrixDisplay(confusion_matrix=cm)
p.plot()
```

```python
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.legend()
```

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
```

```python
import tensorflow as tf
from tensorflow.keras import datasets,regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,BatchNormalization,Dropout
import matplotlib.pyplot as plt
```

RNN

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import Input,Model
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
dataset_path = "SMSSpamCollection"
df = pd.read_csv(dataset_path, sep='\t', header=None, names=['label', 'message'], encoding='utf-8')
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
tokenizer=Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df['message'])
seq=tokenizer.texts_to_sequences(df['message'])
max_length=max(len(x) for x in seq)
pad_seq=pad_sequences(seq,maxlen=100)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(pad_seq,df['label'],test_size=0.2,random_state=42)
```

```python
vocab_length=len(tokenizer.word_index)+1
```

```python
inputs=Input(shape=(100,))
x=Embedding(vocab_length,32)(inputs)
x=SimpleRNN(64,return_sequences=True)(x)
x=Dropout(0.2)(x)
x=SimpleRNN(32)(x)
x=Dropout(0.2)(x)
x=Dense(64,activation='relu')(x)
x=Dropout(0.2)(x)
outputs=Dense(1,activation='sigmoid')(x)
model=Model(inputs=inputs,outputs=outputs)
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
earlystopping=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
history=model.fit(xtrain,ytrain,epochs=10,validation_data=(xtest,ytest),batch_size=32,validation_split=0.2,callbacks=[earlystopping])
```

```python
loss,acc=model.evaluate(xtest,ytest)
acc
```

```python
ypred=(model.predict(xtest)>0.5).astype('int32')
print(accuracy_score(ytest,ypred))
```

```python
def predict_rnn(message):
    seq=tokenizer.texts_to_sequences(message)
    pad_seq=pad_sequences(seq,maxlen=100,padding='post')
    ypred=model.predict(pad_seq)[0][0]
    confi=ypred if ypred>0.5 else 1-ypred
    print(confi)
    return 'spam' if ypred>0.5 else 'ham'
```

```python
print(predict_rnn("Congratulations! You've won a free gift."))
```

LSTM

```python
import tensorflow as tf
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("energydata_complete.csv")
df.head()
```

```python
x=df.drop(columns=['Appliances','date'])
y=df[['Appliances']]
```

```python
sca=MinMaxScaler(feature_range=(0,1))
xscaled=sca.fit_transform(x)
yscaled=sca.fit_transform(y)
```

```python
def create_seq(x,y,length):
    xseq=[]
    yseq=[]
    for i in range(len(x)-length):
        xseq.append(x[i:i+length])
        yseq.append(y[i+length])
    return np.array(xseq),np.array(yseq)
```

```python
sequence_length=10
train_size=int(len(xscaled)*0.8)
```

```python
xtrain_data=xscaled[:train_size]
xtest_data=xscaled[train_size-sequence_length:]
ytrain_data=yscaled[:train_size]
ytest_data=yscaled[train_size-sequence_length:]
```

```python
xtrain,ytrain=create_seq(xtrain_data,ytrain_data,sequence_length)
xtest,ytest=create_seq(xtest_data,ytest_data,sequence_length)
```

```python
print(xtrain.shape,ytrain.shape)
```

```python
model=Sequential([
    LSTM(50,input_shape=(10,27),return_sequences=True),
    Dropout(0.1),
    LSTM(25),
    Dropout(0.1),
    Dense(12),
    Dense(1)
])
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_absolute_error'])
```

```python
earlystopping=tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
```

```python
lr_scheduler=tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',patience=15,min_lr=0.0001)
```

```python
history=model.fit(xtrain,ytrain,batch_size=40,epochs=50,validation_split=0.2,callbacks=[earlystopping,lr_scheduler])
```

```python
train_pred=model.predict(xtrain)
test_pred=model.predict(xtest)
```

```python
train_pred=sca.inverse_transform(train_pred)
test_pred=sca.inverse_transform(test_pred)
ytest_inv=sca.inverse_transform(ytest)
ytrain_inv=sca.inverse_transform(ytrain)
```

```python
def cal_met(actual,pred):
    metrics={
        "RMSE":math.sqrt(mean_squared_error(actual,pred)),
        "MAE":mean_absolute_error(actual,pred),
        "r2_score":r2_score(actual,pred)
    }
    return metrics
```

```python
train_met=cal_met(ytrain_inv,train_pred)
test_met=cal_met(ytest_inv,test_pred)
```

```python
plt.plot(ytrain_inv,label='actual')
plt.plot(train_pred,label='pred')
plt.legend()
```

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
```

```python
metrics_name=['RMSE','MAE','R2']
plt.bar([0-0.1,1-0.1,2-0.1],[train_met['RMSE'],train_met['MAE'],train_met['r2_score']],0.35,label='training')
plt.bar([0+0.1,1+0.1,2+0.1],[test_met['RMSE'],test_met['MAE'],test_met['r2_score']],0.35,label='testing')
plt.legend()
plt.xticks([0,1,2],metrics_name)
```

Autoencoder-Convolutional

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,Flatten,Conv2D,Activation,Input,Reshape,MaxPooling2D,UpSampling2D,BatchNormalization
from tensorflow.keras import datasets
```

```python
noise_factor=0.2
(xtrain,_),(xtest,_)=datasets.cifar10.load_data()
xtrain,xtest=xtrain/255.0,xtest/255.0
xtrain_noise=xtrain.copy()
xtest_noise=xtest.copy()
xtrain_noise=xtrain_noise+noise_factor*np.random.normal(loc=0.0,scale=1.0,size=xtrain.shape)
xtest_noise=xtest_noise+noise_factor*np.random.normal(loc=0.0,scale=1.0,size=xtest.shape)
xtrain_noise=np.clip(xtrain_noise,0.0,1.0)
xtest_noise=np.clip(xtest_noise,0.0,1.0)
```

```python
input_img=Input(shape=(32,32,3))
x=Conv2D(32,(3,3),padding='same')(input_img)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=MaxPooling2D((2,2),padding='same')(x)

x=Conv2D(64,(3,3),padding='same')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=MaxPooling2D((2,2),padding='same')(x)

x=Conv2D(128,(3,3),padding='same')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
encoded=MaxPooling2D((2,2),padding='same')(x)

x=Conv2D(128,(3,3),padding='same')(encoded)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=UpSampling2D((2,2))(x)

x=Conv2D(64,(3,3),padding='same')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=UpSampling2D((2,2))(x)

x=Conv2D(32,(3,3),padding='same')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=UpSampling2D((2,2))(x)

decoded=Conv2D(3,(3,3),padding='same',activation='sigmoid')(x)

autoencoder=Model(input_img,decoded)
encoder=Model(input_img,encoded)

encoded_shape=Input(shape=(4,4,128))
decoded_layers=autoencoder.layers[-13:]
x=encoded_shape
for layers in decoded_layers:
    x=layers(x)
decoder=(encoded_shape,x)
```

```python
autoencoder.compile(optimizer=Adam(learning_rate=0.001),loss='mse')
early_stopping=tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
history=autoencoder.fit(xtrain_noise,xtrain,validation_data=(xtest_noise,xtest),epochs=5,callbacks=[earlystopping],shuffle=True,batch_size=100)
```

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
```

```python
result=autoencoder.predict(xtest[:10])
n=10
plt.figure(figsize=(20,4))
for i in range(10):
    ax=plt.subplot(3,n,i+1)
    plt.imshow(xtest[i])
    plt.title('original')

    ax=plt.subplot(3,n,n+i+1)
    plt.imshow(xtest_noise[i])
    plt.title('noisy')

    ax=plt.subplot(3,n,2*n+i+1)
    plt.imshow(result[i])
    plt.title("reconstruction")
plt.show()
```

Autoencode-MLP

```python
input_shape=(32,32,3)
input_dim=np.prod(input_shape)
encoding_dim=256
input_img=Input(shape=input_shape)
flat_img=Flatten()(input_img)
encoded=Dense(512,activation='relu')(flat_img)
encoded=Dense(encoding_dim,activation='relu')(encoded)
decoded=Dense(512,activation='relu')(encoded)
decoded=Dense(input_dim,activation='relu')(decoded)
output=Reshape(input_shape)(decoded)
autoencoder1=Model(input_img,output)
encoder=Model(input_img,encoded)
encoded_input=Input(shape=(encoding_dim,))
decode_layer=autoencoder1.layers[-3](encoded_input)
decode_layer=autoencoder1.layers[-2](decode_layer)
decode_output=Reshape(input_shape)(decode_layer)
decoder=Model(encoded_input,decode_output)
```

```python
autoencoder1.compile(optimizer=Adam(learning_rate=0.001),loss='mse')
earlystopping1=tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
history1=autoencoder1.fit(xtrain_noise,xtrain,validation_data=(xtest_noise,xtest),batch_size=100,shuffle=True,epochs=5,callbacks=[earlystopping1])
```

```python
plt.plot(history1.history['loss'],label='loss')
plt.plot(history1.history['val_loss'],label='val_loss')
plt.legend()
```
