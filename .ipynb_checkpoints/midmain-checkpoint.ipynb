```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
x=df.data
y=df.target
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense,Dropout
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from keras.utils import to_categorical
```

```python
y=to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = Sequential([Dense(16,activation="relu",input_shape=(xtrain.shape[1],)),Dropout(0.3),Dense(8,activation="relu"),Dropout(0.3),Dense(3,activation="softmax")])
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history=model.fit(xtrain,ytrain,epochs=200,batch_size=32,validation_split=0.2,verbose=1)
```

```python
loss,accuracy=model.evaluate(xtest,ytest)
```

```python
print(loss)
```

```python
print(accuracy)
```

```python
ypred=model.predict(xtest)
```

```python
ypred_class=np.argmax(ypred,axis=1)
ytest_class=np.argmax(ytest,axis=1)
```

```python
cm=confusion_matrix(ytest_class,ypred_class)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
dist=ConfusionMatrixDisplay(confusion_matrix=cm)
dist.plot()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.isna().sum()
```

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
```

```python
df=df.dropna()
```

```python
df.duplicated().sum()
```

```python
print(df.dtypes)
```

```python
enco=LabelEncoder()
df['Sex']=enco.fit_transform(df['Sex'])
df['Embarked']=enco.fit_transform(df['Embarked'])
```

```python
x = df.drop(['Survived'], axis=1)
y = df['Survived']
```

```python
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(xtrain.shape[1])
```

```python
def create_model():
    model = Sequential([Dense(32,activation="relu",input_shape=(7,)),Dense(16,activation="relu"),Dense(1,activation="sigmoid")])
    return model
```

```python
optimizers = {
 "SGD":tf.keras.optimizers.SGD(learning_rate=0.01),
"Momentum":tf.keras.optimizers.SGD(learning_rate=0.01,momentum=0.9),
    "Adam":tf.keras.optimizers.Adam(learning_rate=0.01),
    "RMSprop":tf.keras.optimizers.RMSprop(learning_rate=0.01)

    
}
```

```python
schedules = {
"Step Decay":tf.keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=0.01,decay_steps=10000,decay_rate=0.9,staircase=True),
    "Exponential Decay":tf.keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=0.01,decay_steps=10000,decay_rate=0.9,staircase=False),
    "Piecewise Constand Decay":tf.keras.optimizers.schedules.PiecewiseConstantDecay(boundaries=[5000,10000],values=[0.01,0.005,0.001]),
    "Cosine decay":tf.keras.optimizers.schedules.CosineDecay(initial_learning_rate=0.01,decay_steps=10000,alpha=0.1)

    
}
```

```python
results={}
histories={}

for optname,opt in optimizers.items():
    for schname,sche in schedules.items():
        print(f"Training with {optname} and {schname}...")
        model=create_model()
        model.compile(optimizer=type(opt)(learning_rate=sche),loss="binary_crossentropy",metrics=["accuracy"])
        history=model.fit(xtrain,ytrain,epochs=20,batch_size=32,validation_data=(xtest,ytest),verbose=0)
        results[(optname,schname)]= model.evaluate(xtest,ytest,verbose=0)
        histories[(optname,schname)]=history
        
```

```python
for key, value in results.items():
    print(f"Optimizer: {key[0]}, Schedule: {key[1]} -> Loss: {value[0]:.4f}, Accuracy: {value[1]:.4f}")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1, figsize=(10, 12))
for key,history in histories.items():
    axes[0].plot(history.history['loss'],label = f" {key[0]} - {key[1]}")
    axes[1].plot(history.history['accuracy'] ,label = f" {key[0]} - {key[1]}")
axes[0].legend()
axes[1].legend()
```
